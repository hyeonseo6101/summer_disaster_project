```python
code = pd.read_excel('C:/Users/hyeon/OneDrive/Desktop/2026-1/공모전/데이터분석/데이터/행정구역/법정동코드 조회자료.xlsx',
                     dtype = {'법정동코드' : str})
```

```python
code.head()
```

```python
# 1. 반지하 데이터(df) 텍스트 합치기
df['시군구코드명'] = df['시군구코드명'].str.strip()
df['법정동코드명'] = df['법정동코드명'].str.strip()
df['법정동명'] = df['시군구코드명'] + " " + df['법정동코드명']

# 2. 법정동코드 데이터 공백 제거
code['법정동명'] = code['법정동명'].str.strip()

# 3. 데이터 병합 
final_df = pd.merge(df, code[['법정동명', '법정동코드']], on='법정동명', how='left')
final_df[['법정동명', '법정동코드', '지하층수', '지하면적']].head()
```

```python
basement_counts = final_df.groupby('법정동코드').size().reset_index(name = '반지하_개수')
basement_counts
```

```python
# 지도 데이터 불러오기
gdf = gpd.read_file("C:/Users/hyeon/OneDrive/Desktop/2026-1/공모전/데이터분석/데이터/행정구역/행정구역_법정동_경계/admstr_zone_lgldong_bndry_24.shp", encoding='cp949')
gdf.head()
```

```python
# 앞 8자리만 잘라내어 'merge_key'를 만듦

gdf['merge_key'] = gdf['EMD_CD'].astype(str).str[:8]
basement_counts['merge_key'] = basement_counts['법정동코드'].astype(str).str[:8]
```

```python
# 8자리 'merge_key'를 기준으로 병
merged_gdf = gdf.merge(basement_counts, on='merge_key', how='left')

# 반지하 주택이 하나도 없는 법정동은 NaN이 되므로 0으로 채움
merged_gdf['반지하_개수'] = merged_gdf['반지하_개수'].fillna(0)
```

```python
m = merged_gdf.explore(
    column='반지하_개수',        
    cmap='YlOrRd',                       # 노란색에서 빨간색으로 갈수록 개수가 많음을 의미
    tooltip=['EMD_NM', '반지하_개수'],    # 마우스를 올리면 '동이름'과 '개수'가 뜸
    style_kwds={'weight': 0.5, 'color': 'gray', 'fillOpacity': 0.8}
)

m.save("seoul_basement_map.html")
```
